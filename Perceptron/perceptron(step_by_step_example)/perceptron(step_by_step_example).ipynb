```python
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable
```

```python
def perceptron(x, y, w, b, alpha, theta):
    
    inPut = x
    target = y
    weights = w
    w1 = weights[0]
    w2 = weights[1]
    bias = b
    learning_rate = alpha
    threshold = theta
    
    delta_w = np.zeros(2)  
    delta_b = 0
        
    yIN = bias + inPut[0] * w1 + inPut[1] * w2
            
    if(yIN > threshold):
        Y = 1

    elif((yIN <= threshold) and (yIN >=- threshold)):
        Y = 0

    elif(yIN <=- threshold):
        Y = -1
        
    if(Y != target):
        for j in range(2):
            delta_w[j] = int(target * inPut[j] * learning_rate)
           
        w1 = w1 + delta_w[0]
        w2 = w2 + delta_w[1]
        
        delta_b = target * learning_rate
        bias = bias + delta_b
            
    return yIN, Y, delta_w, delta_b, w1, w2, bias
```

```python
# inPut
x = [[1, 1, 0, 0], [1, 0, 1, 0]]
# target
y = [1, -1, -1, -1]

# learning rate
alpha = 1

# threshold
theta = 0.2

Bias = 1

yIN = np.zeros(40)
Y = np.zeros(40)
delta_w1 = np.zeros(40)
delta_w2 = np.zeros(40)
delta_b = np.zeros(40)
w1 = np.zeros(41)
w1[0] = 0
w2 = np.zeros(41)
w2[0] = 0
b = np.zeros(41)


result = np.zeros(40)

for i in range(10):
    title = PrettyTable([' x1 ', ' x2 ', ' Bias ', ' yIN ', ' y ', ' t ', ' delta w1 ', ' delta w2 ', ' delta b ', ' w1 ', ' w2 ', ' b '])

    for j in range(4):
        result = perceptron([x[0][j], x[1][j]], y[j], [w1[4*i+j], w2[4*i+j]], b[4*i+j], alpha, theta)
        yIN[4*i+j] = result[0]
        Y[4*i+j] = result[1]
        delta_w1[4*i+j] = result[2][0]
        delta_w2[4*i+j] = result[2][1]
        delta_b[4*i+j] = result[3]
        w1[4*i+j+1] = result[4]
        w2[4*i+j+1] = result[5]
        b[4*i+j+1] = result[6]
        title.add_row([x[0][j], x[1][j], Bias, int(yIN[4*i+j]), int(Y[4*i+j]), y[j], int(delta_w1[4*i+j]), int(delta_w2[4*i+j]), int(delta_b[4*i+j]), int(w1[4*i+j+1]), int(w2[4*i+j+1]), int(b[4*i+j+1])])

    print('\nstep ' + str(i+1) + 'th ')
    print(title)
```
