```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def perceptron_for_AND(x, y, bias, weights, learning_rate, threshold, f):
    
    w = weights
    b = bias
    target = y
    alpha = learning_rate
    theta = threshold
    
    for index in range(4):
        
        flag = f
        inPut = [x[0][index], x[1][index]]
        
        while(flag):
            
            flag = 0
            
            yin = b + inPut[0] * w[0] + inPut[1] * w[1]
                
            if(yin > theta):
                y = 1
                    
            elif((yin <= theta) and (yin >=- theta)):
                y = 0
                
            elif(yin <=- theta):
                y = -1
                    
            if(y != target[index]):
                flag = 1
                for j in range(2):
                    w[j] = w[j] + target[index] * inPut[j] * alpha
                        
                b = b + target[index] * alpha
                        
        # display
        
        try:
            m = -(w[0] / w[1])
            c = -b / w[1]
        except ZeroDivisionError:
            m = 0
            c = 0
        s1 = np.linspace(-2,2,100)
        s2 = m * s1 + c + theta
        s3 = m * s1 + c - theta
        
        plt.plot([0, 0, 0], [-2, 0, 2], 'black')
        plt.plot([-2, 0, 2], [0, 0, 0], 'black')
        plt.plot([1, -1, -1], [-1, 1, -1], 'ro')
        plt.plot([1], [1], 'go')
        plt.plot(s1,s2, 'blue')
        plt.plot(s1,s3, 'blue')
        plt.axis([-2, 2, -2, 2])
        
        plt.ylabel('Learning with the ' + str(index) + 'th' + ' educational vector')
        plt.show()
```
