```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hebb_for_OR(x, y, bias, weights):
    
    w = weights
    b = bias
    target = y
    
    for index in range(4):
        
        inPut = [x[0][index], x[1][index]]
        
        for i in range(2):
            w[i] = w[i] + inPut[i] * target[index]
        b = b + target[index]
            
        # display  
        
        try:
            m = -(w[0] / w[1])
            c = -b / w[1]
        except ZeroDivisionError:
            m = 0
            c = 0
            
        s1 = np.linspace(-2,2,100)
        s2 = m*s1+c

        plt.plot([0, 0, 0], [-2, 0, 2], 'black')
        plt.plot([-2, 0, 2], [0, 0, 0], 'black')
        plt.plot([1, 1, -1], [1, -1, 1], 'go')
        plt.plot([-1], [-1], 'ro')
        plt.plot(s1,s2, 'blue')
        plt.axis([-2, 2, -2, 2])
        
        plt.ylabel('Learning with the ' + str(index) + 'th' + ' educational vector')
        plt.show()
```
